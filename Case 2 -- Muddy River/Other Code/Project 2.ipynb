```python
np.arange(0, 12, 0.25)
```

```python
gam = LinearGAM().gridsearch(X, y)
XX = gam.generate_X_grid(term=0, n=500)
```

```python
def calibrate(y):
    from pygam import LinearGAM
    assert len(y) == 13
    y = np.array(y).reshape(-1, 1)
    gam = LinearGAM().gridsearch(np.arange(13).reshape(-1, 1), y)
    return gam.predict(np.arange(0, 12, 0.25))
```

```python
calibrate([2.81, 2.76, 2.73, 3.20, 3.88, 4.45, 4.80, 4.56, 4.13, 3.68, 3.49, 3, 2.81])
```

```python
calibrate([38.26, 39.39, 40.56, 43.23, 56.54, 
                   72.7, 84.19, 76.06, 61.32, 50.75, 46.9, 35, 38.26])
```
